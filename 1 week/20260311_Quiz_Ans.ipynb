```python
# Quiz 1
다음 코드의 출력값과 자료형을 쓰고, 왜 그런 결과가 나오는지 연산 순서를 단계별로 설명하세요.

a = 6
b = 2
result = a / b == a // b * b == 6
print(result)
print(type(result))
```

```python
# Ans 1
출력값은 False, <class 'bool'>

a / b = 6 / 2 = 3.0
a // b = 6 // 2 = 3
a // b * b = 3 * 2 = 6

a / b == a // b * b == 6 → (3.0 == 6) and (6 == 6)

3.0 == 6 → False
False and True → False (즉, 하나라도 False라면 True가 존재하더라도 False 출력)
```

```python
# Quiz 2
다음 코드가 실행될 때, (1) 출력값이 무엇인지, (2) Zero Division Error가 발생하는지, (3) 발생 여부의 이유를 논리 연산 평가 순서로 설명하세요.

x = 4
result = x > 10 and 5 / 0 > 1 or x == 4
print(result)
```

```python
# Ans 2
(1) 출력값은 True
(2) Zero Division Error 발생 X
(3) x > 10 → 4 > 10이 되므로 False, 이로 인해 short-circuit → 5 / 0 > 1은 계산 X
    False or (x == 4)라는 연산이 결과적으로 됐을 때, 4 == 4는 True → 결과적으로 True 출력
```

```python
# Quiz 3
다음 두 코드의 출력값을 각각 쓰고, 두 결과가 다른 이유를 연산자 우선순위 관점에서 설명하세요.

A = {1, 2, 3}
print(A | {3, 4} & {4})

A = {1, 2, 3}
print((A | {3, 4}) & {4})
```

```python
# Ans 3
연산에서 우선순위 연산은 & → | 

{3,4} & {4} → {4}이므로 A |{4}는 {1,2,3} | {4} → {1,2,3,4}
괄호 존재로 A | {3,4} → {1,2,3,4} 계산 후, {1,2,3,4} & {4} → {4}

결과적으로 연산 우선순위와 괄호의 영향 때문에 결과값이 달라짐.
```

```python
# Quiz 5
다음 코드의 출력값을 쓰고, 그 결과가 직관과 다르게 나오는 이유를 설명하세요.

a = 0.1 + 0.2
print(a == 0.3)
print(type(a))
```
