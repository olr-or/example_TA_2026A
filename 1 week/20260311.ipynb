```python
# Example 1

text = "Python3"

result = int(text[-1]) ** 2 + len(text[:4])
# 마지막 문자 "3" → 정수 3 → 3^2=9, "Pyth" 길이 4 → 9+4=13

print(result)
print(type(result))
```

```python
# Example 2

nums = [1, 2, 3]

result = nums[0] + nums[1] * True > nums[2] - False
# True=1, False=0 → 1+2*1=3, 3-0=3 → 3>3 → False

print(result)
print(type(result))
```

```python
# Example 3

s = {1, 2, 3}

result = len((s | {4,5}) - (s & {2,3})) * 2
# 합집합 {1,2,3,4,5}, 교집합 {2,3} → 차집합 {1,4,5} → 길이3 → 3*2=6

print(result)
print(type(result))
```

```python
# Example 4

d = {"x": 2, "y": 5}

result = d["x"] ** d["y"] > 20 and d["y"] % d["x"] == 1
# 2^5=32 → 32>20 True, 5%2=1 → True → True and True

print(result)
print(type(result))
```

```python
# Example 5

text = "12"
nums = [3, 4]

result = int(text[0]) * nums[0] + int(text[1]) * nums[-1]
# 1*3 + 2*4 = 3 + 8 = 11

print(result)
print(type(result))
```
