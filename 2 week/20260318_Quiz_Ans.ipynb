```python
# Quiz 1
다음 코드의 출력값과 자료형을 쓰고, 계산 과정에서 연산자 우선순위가 어떻게 적용되는지 단계별로 설명하세요.

x = 9
y = 4

value = (x % y) ** 2 + x // y * 3
print(value)
print(type(value))
```

```python
# Ans 1
출력값은 7, <class 'int'>

연산자 우선순위: () → ** → *, // → +

우선순위 적용한 계산 과정
(1) x % y = 9 % 4 = 1
(2) (1) ** 2 = 1
(3) x // y = 9 // 4 = 2
(4) 2 * 3 = 6
(5) 1 + 6 = 7
```

```python
# Quiz 2
다음 코드가 실행될 때, (1) 출력값이 무엇인지, (2) 조건문이 어떤 순서로 평가되는지, (3) 어떤 조건이 True/False인지를 설명하세요.

a = 7
b = 5

if a % 2 == 1 and b % 2 == 0 or a + b > 10:
    print("Condition A")
else:
    print("Condition B")
```

```python
# Ans 2
(1) 출력값은 Condition A

(2) 연산자 우선순위: == → and →  or

(3) True / False 조건 평가
     
    i) a % 2 == 1 → 7 % 2 = 1 → True
       b % 2 == 0 → 5 % 2 = 1 → False
    ii) True and False →  False
    iii) a + b > 10 → 7 + 5 = 12 → True
    iV) False or True →  True
```

```python
# Quiz 3
다음 코드의 출력값을 구하고, bool값이 정수처럼 계산되는 이유를 설명하세요.

p = True
q = False

result = p * 4 + q * 6 - 3
print(result)
print(type(result))
```

```python
# Ans 3
출력값은 1, <class 'int'>

* 계산 과정
p = True → 1
q = False → 0

1 * 4 + 0 * 6 - 3 = 4 + 0 - 3 = 1

Python에서 bool은 int의 subclass에 해당하므로 True == 1, False == 0을 의미함.
```

```python
# Quiz 4
다음 코드의 출력값을 쓰고, 조건문이 실행되는 이유를 비교 연산과 논리 연산 순서대로 설명하세요.

x = 10
y = 3

if x / y > x // y and x % y == 1:
    print("Case 1")
else:
    print("Case 2")
```

```python
# Ans 4
출력값은 Case 1

* 계산 과정
(1) x / y = 10 / 3 ≈ 3.333...
    x // y = 10 // 3 = 3
    3.333... > 3 이므로 True
(2) x % y = 10 % 3 = 1 → True
(3) True and True → True
```

```python
# Quiz 5
다음 코드의 출력값과 자료형을 쓰고, 그 결과가 나오는 이유를 자료형 변환(bool → int)과 논리 연산 순서 관점에서 설명하세요.

a = True
b = 3

result = (a + b ** 2 > 10) or (a * b ** 2 - b == 6 and not b % 2)
print(result)
print(type(result))
```

```python
# Ans 5
출력값은 False, <class 'bool'>

* 계산 과정
(1) b ** 2 = 9
(2) a + 9 = True + 9 = 1 + 9 = 10
    10 > 10 → False
(3) a * 9 - 3 = 1 * 9 - 3 = 6
    6 == 6 → True
(4) b % 2 = 3 % 2 = 1
    not 1 → False
(5) True and False → False
(6) False or False → False
```

```python
# Quiz 6
다음 코드의 출력값을 각각 구하고, 두 결과가 다른 이유를 연산자 우선순위와 괄호의 역할 관점에서 설명하세요.

#[First]
S = {2, 4, 6, 8}
print(S - {4, 6} | {1, 3} & {3})

#[Second]
S = {2, 4, 6, 8}
print((S - {4, 6} | {1, 3}) & {3})
```

```python
# Ans 6
연산자 우선순위가 & → - → |

[First]
(1) {1, 3} & {3} = {3}
(2) S - {4, 6} = {2, 8}
(3) 출력은 결론적으로 {2, 3, 8}

[Second]
(1) S - {4, 6} = {2, 8}
    {2, 8} | {1, 3} = {1, 2, 3, 8}
(2) {1, 2, 3, 8} & {3} = {3}
```
