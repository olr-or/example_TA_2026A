```python
# 응용 예제 #
```

```python
# 1

x = 5
y = 2

result = x ** y + y ** x
print(result)
```

```python
# 2

a = 8
b = 3

result = a // b * b + a % b
print(result)
```

```python
# 3

num = 15

if num % 3 == 0 and num % 5 == 0:
    print("3과 5의 공배수")
```

```python
# 4

a = 4
b = 6

if a ** 2 + b ** 2 > 40:
    print("조건 만족")
else:
    print("조건 불만족")
```

```python
# 5

x = 7

if x % 2 == 1 and x > 5:
    print("홀수이면서 5보다 큼")
```

```python
# 6

a = 10
b = 4

result = (a / b) > (a // b)
print(result)
```

```python
# 7

x = 6
y = 3

if x / y == x // y:
    print("같은 결과")
else:
    print("다른 결과")
```

```python
# 8

a = True
b = False

result = a * 10 + b * 5 + 2
print(result)
```

```python
# 9

num = 12

if num % 4 == 0 or num % 5 == 0:
    print("조건 충족")
```

```python
# 10

a = 5
b = 2

result = (a + b) ** 2 - a * b
print(result)
```

```python
# 11

x = 5
y = 2

if (x ** y > y ** x) or (x + y) % 2 == 0:
    print("조건 만족")
else:
    print("조건 불만족")
```

```python
# 12

x = 6
y = 2

if (x / y == x // y) and (x % y == 0):
    print("정수 나눗셈")
else:
    print("실수 결과")
```
