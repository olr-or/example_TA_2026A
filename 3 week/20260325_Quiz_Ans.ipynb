```python
# Quiz 1
다음 조건에 따라 (1) 코드를 작성하고, (2) 코드 기반으로 아래 내용을 출력한 뒤, (3) is 연산자를 활용하여 a is b와 t1 is t2를 비교하고, (4) 리스트와 튜플의 차이를 설명하세요.

[코드 작성 조건]
* 리스트 a = [10, 20], 튜플 t1 = (10, 20)을 생성한다.
* 딕셔너리 d1 = {"x": a, "y": t1}를 만든다.
* a를 복사하여 b, t1을 복사하여 t2에 저장한다.
* 이후 다음 연산을 수행한다:
    - a += [30] 
    - t1 += (30,)

[출력]
* a, b * t1, t2 * d1
```

```python
# Ans 1
a = [10, 20] t1 = (10, 20)

d1 = {“x”: a, “y”: t1}

b = a t2 = t1

a += [30] t1 += (30,)

print(“a =”, a) print(“b =”, b) print(“t1 =”, t1) print(“t2 =”, t2)
print(“d1 =”, d1)

print(“a is b ->”, a is b) print(“t1 is t2 ->”, t1 is t2)

리스트는 가변 객체이므로 a를 수정하면 b에도 반영 O. 
튜플은 불변 객체이므로 t1 변경 시 새로운 객체가 생성.
```

```python
# Quiz 2
정수 리스트 nums = [5, -2, 0, 11, -7, 8] 가 있을 때, 각 원소에 대해 다음 규칙을 적용하여 새 리스트 result를 만드세요.

[규칙 및 조건]
* 양수이면서 짝수 → "PE"
* 양수이면서 홀수 → "PO"
* 0 → "Z"
* 음수 → "N"
* list comprehension을 사용할 것
```

```python
# Ans 2
nums = [5, -2, 0, 11, -7, 8]

result = [ “PE” if n > 0 and n % 2 == 0 else “PO” if n > 0 and n % 2 ==
1 else “Z” if n == 0 else “N” for n in nums]

print(result)
```

```python
# Quiz 3
문자열 text = "machinelearning" 일 때, 다음 조건을 만족하는 코드를 작성하세요.

[조건]
* 문자열을 두 부분으로 나눈다:
- 앞부분: "machine"
- 뒷부분: "learning"
* 뒷부분이 "learning"이면 "AI"로 변경 
* 최종 문자열을 모두 대문자로 변환
```

```python
# Ans 3
text = “machinelearning”

front = text[:7] back = text[7:]

if back == “learning”: back = “AI”

result = (front + back).upper()
print(result)
```

```python
# Quiz 4
정수 score를 입력받아 다음 기준으로 등급을 출력하세요. (딕셔너리를 활용할 것)

* 90 이상 → "A" 
* 80 이상 → "B" 
* 70 이상 → "C" 
* 그 외 → "F"

출력 예시: 85점은 B등급입니다.
```

```python
# Ans 4
score = int(input(“점수를 입력하세요:”))

grade_dict = { 9: “A”, 8: “B”, 7: “C” }
grade = grade_dict.get(score // 10, “F”)

print(f”{score}점은 {grade} 등급입니다.”)
```

```python
# Quiz 5
다음 2차원 리스트가 있을 때, 

matrix = [
    [2, 7, 4],
    [9, 3, 8],
    [1, 6, 5]
]

아래 조건을 만족하는 값만 추출하여 리스트 selected를 만드세요.

* 홀수이면서 5보다 큰 수
* 리스트를 내포하면서 조건 필터링을 사용할 것
```

```python
# Ans 5
matrix = [ [2, 7, 4], [9, 3, 8], [1, 6, 5]]
selected = [x for row in matrix for x in row if x % 2 == 1 and x > 5]

print(selected)
```
