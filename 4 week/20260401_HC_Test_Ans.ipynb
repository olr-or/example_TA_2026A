```python
# Q1) Digit Frequency Counter

n = input("정수를 입력하세요: ")

freq = {}
i = 0
while i < len(n):
    digit = int(n[i])
    if digit in freq:
        freq[digit] = freq[digit] + 1
    else:
        freq[digit] = 1
    i = i + 1

print(freq)
```

```python
# Q2) List Rotation

data = input("숫자들을 공백으로 입력하세요: ")
k = int(input("k 값을 입력하세요: "))

lst = data.split()

n = len(lst)

if n > 0:
    k = k % n

result = []
i = n - k
while i < n:
    result.append(lst[i])
    i = i + 1

i = 0
while i < n - k:
    result.append(lst[i])
    i = i + 1

print(result)
```

```python
# Q4) Custom Sorting Without Built-in Functions

data = input("숫자들을 공백으로 입력하세요: ")
arr_str = data.split()

arr = []
i = 0
while i < len(arr_str):
    arr.append(int(arr_str[i]))
    i = i + 1

n = len(arr)

i = 0
while i < n - 1:
    min_idx = i
    j = i + 1
    while j < n:
        if arr[j] < arr[min_idx]:
            min_idx = j
        j = j + 1

    temp = arr[i]
    arr[i] = arr[min_idx]
    arr[min_idx] = temp

    i = i + 1

print(arr)
```

```python
# Q5) Find the Second Largest Number

data = input("숫자들을 공백으로 입력하세요: ")
arr_str = data.split()

arr = []
i = 0
while i < len(arr_str):
    arr.append(int(arr_str[i]))
    i = i + 1

largest = arr[0]
second = arr[0]

i = 1
while i < len(arr):
    if arr[i] > largest:
        second = largest
        largest = arr[i]
    elif arr[i] > second and arr[i] != largest:
        second = arr[i]
    i = i + 1

print(second)
```

```python
# Q6) Find Second Minimum and Third Maximum in a List

data = input("숫자들을 공백으로 입력하세요: ")
arr_str = data.split()

arr = []
i = 0
while i < len(arr_str):
    num = int(arr_str[i])
    if num not in arr:
        arr.append(num)
    i = i + 1

if len(arr) < 3:
    print("Invalid input")
else:
    smallest = arr[0]
    second_smallest = arr[0]

    i = 1
    while i < len(arr):
        if arr[i] < smallest:
            second_smallest = smallest
            smallest = arr[i]
        elif (arr[i] < second_smallest or second_smallest == smallest) and arr[i] != smallest:
            second_smallest = arr[i]
        i = i + 1

    largest = arr[0]
    second_largest = arr[0]
    third_largest = arr[0]

    i = 1
    while i < len(arr):
        if arr[i] > largest:
            third_largest = second_largest
            second_largest = largest
            largest = arr[i]
        elif arr[i] != largest and (arr[i] > second_largest or second_largest == largest):
            third_largest = second_largest
            second_largest = arr[i]
        elif arr[i] != largest and arr[i] != second_largest and (arr[i] > third_largest or third_largest == second_largest or third_largest == largest):
            third_largest = arr[i]
        i = i + 1

    print("Second Min:", second_smallest)
    print("Third Max:", third_largest)
```

```python
# Q7) Twin Prime Checker

n = int(input("정수를 입력하세요: "))

def is_prime(x):
    if x < 2:
        return False

    i = 2
    while i < x:
        if x % i == 0:
            return False
        i = i + 1

    return True

if is_prime(n) and (is_prime(n - 2) or is_prime(n + 2)):
    print("N is a twin prime.")
else:
    print("N is not a twin prime.")
```

```python
# Q8) Conditional Summation with Divisibility and Digit Constraints

n = int(input("N을 입력하세요: "))

sum_div4 = 0
sum_digit3 = 0
sum_both = 0

num = 1
while num <= n:
    temp = num
    digit_sum = 0

    while temp > 0:
        digit_sum = digit_sum + (temp % 10)
        temp = temp // 10

    if num % 4 == 0:
        sum_div4 = sum_div4 + num

    if digit_sum % 3 == 0:
        sum_digit3 = sum_digit3 + num

    if num % 4 == 0 and digit_sum % 3 == 0:
        sum_both = sum_both + num

    num = num + 1

print("Divisible by 4:", sum_div4)
print("Digit sum divisible by 3:", sum_digit3)
print("Both conditions:", sum_both)
```

```python
# Q9) Frequency-Based Sorting of Matrix Elements

data = input("2x2 행렬 원소 4개를 공백으로 입력하세요: ")
arr_str = data.split()

arr = []
i = 0
while i < len(arr_str):
    arr.append(int(arr_str[i]))
    i = i + 1

freq = {}
i = 0
while i < len(arr):
    if arr[i] in freq:
        freq[arr[i]] = freq[arr[i]] + 1
    else:
        freq[arr[i]] = 1
    i = i + 1

n = len(arr)
i = 0
while i < n - 1:
    j = 0
    while j < n - 1 - i:
        left = arr[j]
        right = arr[j + 1]

        if freq[left] > freq[right]:
            temp = arr[j]
            arr[j] = arr[j + 1]
            arr[j + 1] = temp
        elif freq[left] == freq[right]:
            if left > right:
                temp = arr[j]
                arr[j] = arr[j + 1]
                arr[j + 1] = temp

        j = j + 1
    i = i + 1

print(arr)
```

```python
# Q10) Palindrome Checker with Explanation

s = input("문자열을 입력하세요: ")

reversed_s = ""
i = len(s) - 1
while i >= 0:
    reversed_s = reversed_s + s[i]
    i = i - 1

if s == reversed_s:
    print("Palindrome")
else:
    print("Not Palindrome")
```
